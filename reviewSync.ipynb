```python
import asyncio
from playwright.async_api import async_playwright
from bs4 import BeautifulSoup
import nest_asyncio
import requests
from pymongo import MongoClient

# Apply nest_asyncio to patch the event loop
nest_asyncio.apply()

class Review:
    def __init__(self, reviewer_name, profession, review_content):
        self.reviewer_name = reviewer_name
        self.profession = profession
        self.review_content = review_content

    def to_dict(self):
        return {
            "reviewer_name": self.reviewer_name,
            "profession": self.profession,
            "review_content": self.review_content
        }

    def __repr__(self):
        return f"Review(reviewer_name='{self.reviewer_name}', profession='{self.profession}', review_content='{self.review_content}')"

class Movie:
    def __init__(self, movie_name, image_url):
        self.movie_name = movie_name
        self.image_url = image_url
        self.scores = {'critics': None, 'audience': None}
        self.reviews = []

    def add_review(self, review):
        self.reviews.append(review)

    def set_scores(self, critics_score, audience_score):
        self.scores['critics'] = critics_score
        self.scores['audience'] = audience_score

    def to_dict(self):
        return {
            "movie_name": self.movie_name,
            "image_url": self.image_url,
            "scores": self.scores,
            "reviews": [review.to_dict() for review in self.reviews]
        }

    def __repr__(self):
        return f"Movie(movie_name='{self.movie_name}', image_url='{self.image_url}', scores={self.scores}, reviews={self.reviews})"

def fetch_reviews_from_view_all(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    reviews = []
    review_cards = soup.select('review-card-critic')

    for card in review_cards:
        reviewer_name = card.select_one('rt-link[slot="displayName"] rt-text').text.strip()
        profession = card.select_one('rt-link[slot="publicationName"] rt-text').text.strip()
        review_content = card.select_one('drawer-more[slot="reviewQuote"] rt-text').text.strip()

        review = Review(reviewer_name, profession, review_content)
        reviews.append(review)

    return reviews

async def fetch_movie_data(max_movies=100):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        # Navigate to the target URL
        url = 'https://www.rottentomatoes.com/browse/movies_at_home/sort:popular'
        await page.goto(url)

        async def extract_movies():
            content = await page.content()
            soup = BeautifulSoup(content, 'html.parser')
            movie_cards = soup.select('div[data-qa="discovery-media-list-item"]')
            movies_info = []
            for card in movie_cards:
                title_element = card.select_one('.p--small')
                title = title_element.get_text(strip=True)
                critics_score_element = card.select_one('rt-text[slot="criticsScore"]')
                critics_score = critics_score_element.get_text(strip=True) if critics_score_element else None
                audience_score_element = card.select_one('rt-text[slot="audienceScore"]')
                audience_score = audience_score_element.get_text(strip=True) if audience_score_element else None
                link_element = card.select_one('a[data-qa="discovery-media-list-item-caption"]')
                link = link_element.get('href') if link_element else None
                image_element = card.select_one('rt-img[slot="image"]')
                image_url = image_element['src'] if image_element and 'src' in image_element.attrs else None
                movies_info.append({
                    'title': title,
                    'critics_score': critics_score,
                    'audience_score': audience_score,
                    'link': f"https://www.rottentomatoes.com{link}" if link else None,
                    'image_url': image_url
                })
            return movies_info

        movies_info = await extract_movies()
        total_movies = len(movies_info)

        while total_movies < max_movies:
            try:
                load_more_button_selector = 'button[data-qa="dlp-load-more-button"]'
                await page.wait_for_selector(load_more_button_selector, timeout=5000)
                await page.click(load_more_button_selector)
                await page.wait_for_timeout(5000)
                more_movies_info = await extract_movies()
                movies_info.extend(more_movies_info)
                total_movies = len(movies_info)
            except:
                break

        await browser.close()

    movies = []
    for movie_info in movies_info:
        movie = Movie(movie_info['title'], movie_info['image_url'])
        movie.set_scores(movie_info['critics_score'], movie_info['audience_score'])
        if movie_info['link']:
            response = requests.get(movie_info['link'])
            soup = BeautifulSoup(response.text, 'html.parser')
            review_cards = soup.select('review-card-critic')
            for card in review_cards:
                reviewer_name = card.select_one('rt-link[slot="displayName"] rt-text').text.strip()
                profession = card.select_one('rt-link[slot="publicationName"] rt-text').text.strip()
                review_content = card.select_one('drawer-more[slot="reviewQuote"] rt-text').text.strip()
                review = Review(reviewer_name, profession, review_content)
                movie.add_review(review)
            view_all_link = soup.select_one('.header-wrap rt-link[href]')
            if view_all_link:
                view_all_url = view_all_link['href']
                full_reviews_url = f"https://www.rottentomatoes.com{view_all_url}" if view_all_url.startswith('/') else view_all_url
                full_reviews = fetch_reviews_from_view_all(full_reviews_url)
                for review in full_reviews:
                    movie.add_review(review)
        movies.append(movie)

    return movies

# def upload_to_mongodb(movies, db_name='movie_reviews', collection_name='movies'):
#     client = MongoClient('mongodb+srv://review123:review123@cluster0.73f4jpp.mongodb.net/')  # Change the connection string if necessary
#     db = client[db_name]
#     collection = db[collection_name]
#     for movie in movies:
#         if not collection.find_one({'movie_name': movie.movie_name}):
#             collection.insert_one(movie.to_dict())
#     print(f"Uploaded {len(movies)} movies to MongoDB.")

def upload_to_mongodb(movies, db_name='movie_reviews', collection_name='movies'):
    client = MongoClient('mongodb+srv://review123:review123@cluster0.73f4jpp.mongodb.net/')  # Change the connection string if necessary
    db = client[db_name]
    collection = db[collection_name]
    for movie in movies:
        movie_json = movie.to_dict()  # Convert Movie object to JSON format
        if not collection.find_one({'movie_name': movie_json['movie_name']}):
            collection.insert_one(movie_json)
    print(f"Uploaded {len(movies)} movies to MongoDB.")


async def main():
    movies = await fetch_movie_data(max_movies=200)  # Adjust the number of movies as needed
    upload_to_mongodb(movies)

asyncio.get_event_loop().run_until_complete(main())
```

```python
import asyncio
from playwright.async_api import async_playwright, TimeoutError as PlaywrightTimeoutError
from bs4 import BeautifulSoup
import nest_asyncio
from pymongo import MongoClient
import re

# Apply nest_asyncio to patch the event loop
nest_asyncio.apply()

class Movie:
    def __init__(self, title, image_url, scores, link, popular_reviews):
        self.title = title
        self.image_url = image_url
        self.scores = scores
        self.link = link
        self.popular_reviews = popular_reviews

    def to_dict(self):
        return {
            "title": self.title,
            "image_url": self.image_url,
            "scores": self.scores,
            "link": self.link,
            "popular_reviews": self.popular_reviews
        }

    def __repr__(self):
        return f"Movie(title='{self.title}', image_url='{self.image_url}', scores={self.scores}, link='{self.link}', popular_reviews={self.popular_reviews})"

async def fetch_movie_data_from_letterboxd(max_movies=100):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        async def navigate_to_page(url, retries=3, timeout=60000):
            for attempt in range(retries):
                try:
                    await page.goto(url, timeout=timeout)
                    await page.wait_for_selector('li.poster-container', timeout=timeout)
                    return True
                except PlaywrightTimeoutError:
                    print(f"TimeoutError: Retrying navigation to {url} (Attempt {attempt + 1}/{retries})")
            return False

        async def extract_popular_reviews(movie_link):
            await page.goto(movie_link)
            await page.wait_for_selector('section#popular-reviews', timeout=30000)
            content = await page.content()
            soup = BeautifulSoup(content, 'html.parser')

            review_items = soup.select('li.film-detail')
            reviews = []
            for item in review_items:
                reviewer_name = item.select_one('strong.name').get_text(strip=True) if item.select_one('strong.name') else None
                review_content = item.select_one('div.body-text').get_text(strip=True) if item.select_one('div.body-text') else None
                reviews.append({'reviewer_name': reviewer_name, 'review_content': review_content})
            return reviews

        url = 'https://letterboxd.com/films/popular/this/month/'
        if not await navigate_to_page(url):
            print("Failed to navigate to the page after retries.")
            await browser.close()
            return []

        async def extract_movies():
            content = await page.content()
            soup = BeautifulSoup(content, 'html.parser')
            movie_cards = soup.select('li.poster-container')
            movies_info = []
            for card in movie_cards:
                title_element = card.select_one('.frame-title')
                title = title_element.get_text(strip=True) if title_element else "Unknown Title"
                title = re.sub(r'\s*\(\d{4}\)$', '', title)
                rating = float(card.get('data-average-rating', 0))
                critic_score = int(rating * 20)
                audience_score = int(critic_score - critic_score * 0.05)
                link = card.select_one('a.frame').get('href') if card.select_one('a.frame') else None
                image_element = card.select_one('img')
                image_url = image_element.get('src') if image_element else None

                popular_reviews = await extract_popular_reviews(f"https://letterboxd.com{link}") if link else []

                movies_info.append({
                    'title': title,
                    'scores': {'critics': critic_score, 'audience': audience_score},
                    'link': f"https://letterboxd.com{link}" if link else None,
                    'image_url': image_url,
                    'popular_reviews': popular_reviews
                })
            return movies_info

        movies_info = await extract_movies()
        total_movies = len(movies_info)

        while total_movies < max_movies:
            try:
                load_more_button_selector = '.load-more a'
                await page.wait_for_selector(load_more_button_selector, timeout=10000)
                await page.click(load_more_button_selector)
                await page.wait_for_timeout(5000)
                more_movies_info = await extract_movies()
                movies_info.extend(more_movies_info)
                total_movies = len(movies_info)
            except PlaywrightTimeoutError:
                break

        await browser.close()

    movies = [Movie(movie_info['title'], movie_info['image_url'], movie_info['scores'], movie_info['link'], movie_info['popular_reviews']) for movie_info in movies_info]
    return movies

def upload_to_mongodb(movies, db_name='movie_reviews', collection_name='letterboxd_movies'):
    client = MongoClient('mongodb+srv://review123:review123@cluster0.73f4jpp.mongodb.net/')
    db = client[db_name]
    collection = db[collection_name]
    for movie in movies:
        if not collection.find_one({'title': movie.title}):
            collection.insert_one(movie.to_dict())
    print(f"Uploaded {len(movies)} movies to MongoDB.")

async def main():
    movies = await fetch_movie_data_from_letterboxd(max_movies=100)
    upload_to_mongodb(movies)
    for movie in movies:
        print(movie)

asyncio.get_event_loop().run_until_complete(main())
```
